```python
import pandas as pd
```

```python
df_coaches = pd.read_csv(".\male_coaches.csv")
df_teams = pd.read_csv(".\male_teams.csv")
```

```python
df_coaches = df_coaches[["coach_id","long_name"]].astype(str)
df_coaches
```

```python
df_teams = df_teams[["coach_id","team_name"]].astype(str)
df_teams["coach_id"] = df_teams.apply(lambda x: str(x.coach_id.split(".")[0]),axis=1)
df_teams
```

```python
df_teams['coach_id']=df_teams['coach_id'].astype(str)
df_coaches['coach_id']=df_coaches['coach_id'].astype(str)

merged_df = pd.merge(df_teams, df_coaches, on="coach_id", how="inner")
merged_df
```

```python
merged_df = merged_df.drop_duplicates()[["team_name","long_name"]]
merged_df
```

```python
merged_df.to_csv("team_coach.csv",index=False)
```
